## Question 1 — Safety equipment vs fatality

### (a) Data and notation

\[
\begin{array}{c|cc|c}
\text{Safety equipment} & \text{Fatal} & \text{Non-fatal} & \text{Row total}\\ \hline
\text{None} & a = 189 & b = 10{,}843 & n_1 = a + b = 189 + 10{,}843 = 11{,}032 \\
\text{Seat belt} & c = 104 & d = 10{,}933 & n_2 = c + d = 104 + 10{,}933 = 11{,}037 \\
\hline
\text{Column totals} & a + c = 293 & b + d = 21{,}776 & N = n_1 + n_2 = 22{,}069
\end{array}
\]

---

### (b) Sample proportions

\[
\hat{p}_1 = \frac{a}{n_1} = \frac{189}{11{,}032} = 0.017132, \quad (1.7132\%)
\]

\[
\hat{p}_2 = \frac{c}{n_2} = \frac{104}{11{,}037} = 0.009423, \quad (0.9423\%)
\]

---

### (c) Relative risk (RR)

\[
RR = \frac{\hat{p}_1}{\hat{p}_2} 
= \frac{0.017132}{0.009423} 
= 1.818
\]

**Interpretation:** The risk of a fatal crash is **1.82× higher** for drivers without safety equipment.

---

### (d) Odds ratio (OR)

\[
\text{odds}_1 = \frac{a}{b} = \frac{189}{10{,}843} = 0.01743, \quad 
\text{odds}_2 = \frac{c}{d} = \frac{104}{10{,}933} = 0.00951
\]

\[
OR = \frac{\text{odds}_1}{\text{odds}_2} 
= \frac{189 \times 10{,}933}{10{,}843 \times 104} 
= 1.832
\]

**Interpretation:** The odds of fatal injury are **1.83× higher** for drivers without safety equipment.

---

### (e) Absolute risk difference (ARR)

\[
\Delta = \hat{p}_1 - \hat{p}_2 
= 0.017132 - 0.009423 
= 0.007709 \; (\text{or } 0.7709\%)
\]

**Interpretation:** Not wearing safety equipment increases fatality risk by **0.77 percentage points**.

---

### (f) Number Needed to Treat (NNT)

\[
NNT = \frac{1}{\Delta} = \frac{1}{0.007709} \approx 129.7
\]

**Interpretation:** Roughly **130 people** would need to wear seat belts to prevent one fatality.

---

### (g) Confidence intervals

#### (i) Risk difference (Normal approximation)

\[
SE_\Delta = \sqrt{\frac{\hat{p}_1(1-\hat{p}_1)}{n_1} + \frac{\hat{p}_2(1-\hat{p}_2)}{n_2}}
= 0.001540
\]

\[
95\%~CI = \Delta \pm 1.96(SE_\Delta)
= 0.007709 \pm 1.96(0.001540)
= (0.00469,~0.01073)
\]

---

#### (ii) Relative risk (log scale)

\[
SE_{\ln(RR)} = \sqrt{\frac{1-\hat{p}_1}{n_1\hat{p}_1} + \frac{1-\hat{p}_2}{n_2\hat{p}_2}} 
= 0.1213
\]

\[
95\%~CI_{RR} = \exp\left[\ln(1.818) \pm 1.96(0.1213)\right] 
= (1.43,~2.31)
\]

---

#### (iii) Odds ratio (log scale)

\[
SE_{\ln(OR)} = \sqrt{\frac{1}{a}+\frac{1}{b}+\frac{1}{c}+\frac{1}{d}} 
= 0.1228
\]

\[
95\%~CI_{OR} = \exp\left[\ln(1.832) \pm 1.96(0.1228)\right] 
= (1.44,~2.33)
\]

---

### (h) Hypothesis test (two-proportion z-test)

\[
H_0: p_1 = p_2, \quad H_A: p_1 \neq p_2
\]

\[
\hat{p}_{pooled} = \frac{a+c}{n_1+n_2} = \frac{293}{22{,}069} = 0.013276
\]

\[
SE_{pooled} = \sqrt{\hat{p}_{pooled}(1-\hat{p}_{pooled})\left(\frac{1}{n_1}+\frac{1}{n_2}\right)} = 0.001540
\]

\[
z = \frac{\hat{p}_1 - \hat{p}_2}{SE_{pooled}} = \frac{0.007709}{0.001540} = 5.00
\]

\[
p\text{-value} = 2(1 - \Phi(|5.00|)) \approx 5.6\times10^{-7}
\]

**Conclusion:** Very strong evidence that seat belt use is associated with fewer fatalities.

## Question 2 — Oracon use and endometrial cancer

### (a) Data

\[
\begin{array}{c|cc|c}
\text{Outcome} & \text{Exposed (Oracon)} & \text{Unexposed} & \text{Row total}\\ \hline
\text{Cases} & a = 6 & b = 111 & 117 \\
\text{Controls} & c = 8 & d = 387 & 395 \\
\hline
\text{Column totals} & 14 & 498 & 512
\end{array}
\]

---

### (b) Odds ratio and confidence interval

\[
OR = \frac{a \times d}{b \times c} = \frac{6 \times 387}{111 \times 8} = 2.615
\]

\[
SE_{\ln(OR)} = \sqrt{\frac{1}{a}+\frac{1}{b}+\frac{1}{c}+\frac{1}{d}} = 0.551
\]

\[
95\%~CI_{OR} = \exp\left[\ln(2.615) \pm 1.96(0.551)\right] = (0.89,~7.69)
\]

**Fisher’s exact two-sided p-value:** \(p = 0.0999\)

**Interpretation:**  
The odds of Oracon use among cases appear higher (OR≈2.6), but the confidence interval includes 1.  
At α = 0.05, the evidence is **not statistically significant**.  
Small cell counts yield wide uncertainty and low power.

```python
import math

# Q1 data
a,b,c,d = 189,10843,104,10933
n1,n2 = a+b, c+d
p1,p2 = a/n1, c/n2

RR = p1/p2
OR = (a*d)/(b*c)
diff = p1 - p2

SE_diff = math.sqrt(p1*(1-p1)/n1 + p2*(1-p2)/n2)
ci_diff = (diff - 1.96*SE_diff, diff + 1.96*SE_diff)

SE_logRR = math.sqrt((1-p1)/(n1*p1) + (1-p2)/(n2*p2))
ci_RR = (math.exp(math.log(RR)-1.96*SE_logRR), math.exp(math.log(RR)+1.96*SE_logRR))

SE_logOR = math.sqrt(1/a + 1/b + 1/c + 1/d)
ci_OR = (math.exp(math.log(OR)-1.96*SE_logOR), math.exp(math.log(OR)+1.96*SE_logOR))

print("Q1 — Seatbelt Analysis")
print("RR =", RR, "CI:", ci_RR)
print("OR =", OR, "CI:", ci_OR)
print("Risk diff =", diff, "CI:", ci_diff)

# Q2 data
a2,b2,c2,d2 = 6,111,8,387
OR2 = (a2*d2)/(b2*c2)
SE_logOR2 = math.sqrt(1/a2 + 1/b2 + 1/c2 + 1/d2)
ci_OR2 = (math.exp(math.log(OR2)-1.96*SE_logOR2), math.exp(math.log(OR2)+1.96*SE_logOR2))

print("\nQ2 — Oracon Case-Control")
print("OR =", OR2, "CI:", ci_OR2)
